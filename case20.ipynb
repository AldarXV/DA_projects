1. Следует удалить столбцы DiscountAmount, OrderNumber, так как они неинформативны.

```python
df=df.drop(['DiscountAmount', 'OrderNumber', 'RetailPrice', 'UnitPrice', 'ABC', 'XYZ'], axis=1)
```

#Регрессия

```python
df['Year']=df['OrderDate'].dt.year
df['Month']=df['OrderDate'].dt.month
```

```python
df.head()
```

```python
df=df.drop('OrderDate', axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
df1 = df[['ProductName','ProductCategory','ItemGroup','KitType','Channels','Demographic', 'RegionName', 'StateName','SaleAmount', 'Year',	'Month']]
df1=df1.groupby(['ProductName','ProductCategory','ItemGroup', 'KitType','Channels','Demographic', 'RegionName', 'StateName', 'Year','Month'], as_index=False ).mean()
```

```python
df1_dummies = pd.get_dummies(df1)
df1_dummies
```

```python
df1_dummies=df1_dummies.sort_values(by=['Year', 'Month'])
df1_dummies
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1_dummies.drop(columns='SaleAmount').to_numpy(), df1_dummies['SaleAmount'].to_numpy(), test_size = 0.3, random_state = 1)
```

```python
gradient_boosted = GradientBoostingRegressor(random_state=2)
gradient_boosted.fit(X_train, y_train)
model_pred = gradient_boosted.predict(X_test)
print('mae= ', np.mean(abs(model_pred- y_test)))
```

Значение mae=4.953 больше полученного в Knime (2.899).

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
df2 = df[['ProductName','ProductCategory','ItemGroup','KitType','Channels','Demographic', 'RegionName', 'StateName', 'ABC-XYZ', 'OrderProcessingSpeed']]
```

```python
df2_dummies = pd.get_dummies(df2)
```

```python
df2_dummies.columns
```

```python
df2_dummies=df2_dummies.drop('OrderProcessingSpeed_Expired', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df2_dummies.drop(columns='OrderProcessingSpeed_OnTime').to_numpy(), df2_dummies['OrderProcessingSpeed_OnTime'].to_numpy(), test_size = 0.3, random_state = 1)
```

```python
logReg = LogisticRegression(max_iter=300)
logReg.fit(X_train, y_train)
print('Точность')
print('Логистическая регрессия: ', round(logReg.score(X_test, y_test), 4))

random_forest = RandomForestClassifier(random_state=1)
random_forest.fit(X_train, y_train)
print('Random forest', round(random_forest.score(X_test, y_test), 4))
```
